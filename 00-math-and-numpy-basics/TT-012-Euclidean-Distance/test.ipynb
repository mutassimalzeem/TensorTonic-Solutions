```python
Input: x = [3,4], y = [0,0]

Output: 5.0

√((3-0)² + (4-0)²) = √(9 + 16) = √25 = 5
```

```python
import numpy as np

def euclidean_distance(x, y):

    x = np.array(x)
    y = np.array(y)

    z = np.sum((x - y)**2)
    return np.sqrt(z)
euclidean_distance(x = [3, 4], y = [0, 0])
```
